```python
import matplotlib.pyplot as plt
x=["Py","c","c++","java"]
y=[90,70,60,80]
plt.xlabel("language",fontsize=10)
plt.ylabel("value",fontsize=10)
plt.title("Popularity",fontsize=10)
plt.bar(x,y,width=0.5,color="m")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["Py","c","c++","java"]
y=[90,70,60,80]
plt.xlabel("language",fontsize=10)
plt.ylabel("value",fontsize=10)
plt.title("Popularity",fontsize=10)
c=["y","b","g","m"]
plt.bar(x,y,width=0.1,color=c)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["Py","c","c++","java"]
y=[90,70,60,80]
plt.xlabel("language",fontsize=10)
plt.ylabel("value",fontsize=10)
plt.title("Popularity",fontsize=10)
c=["y","b","g","m"]
plt.bar(x,y,width=0.5,color=c,edgecolor="r",linewidth=10,linestyle=":",alpha=0.4)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["Py","c","c++","java"]
y=[90,70,60,80]
plt.xlabel("language",fontsize=10)
plt.ylabel("value",fontsize=10)
plt.title("Popularity",fontsize=10)

plt.bar(x,y,width=0.5,color="r",edgecolor="r",label="popularity")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["Py","c","c++","java"]
y=[90,70,60,80]
z=[20,30,40,50]
plt.xlabel("language",fontsize=10)
plt.ylabel("value",fontsize=10)
plt.title("Popularity",fontsize=10)
plt.bar(x,y,width=0.5,color="r",edgecolor="r",label="popularity")
plt.bar(x,z,width=0.5,color="g",edgecolor="r",label="popularity1")
plt.legend()
plt.show()
```
