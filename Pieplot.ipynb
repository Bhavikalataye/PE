```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c","c++","java","Py"]
ex=[0.1,0.0,0.0,0.0]
c=["r","y","m","g"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",radius=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c","c++","java","Py"]
ex=[0.3,0.0,0.0,0.0]
c=["r","y","m","g"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",radius=1)
plt.title("Language")
plt.legend(loc=2)
plt.show()
```
